```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("mean_channel_data.csv")
df
```

```python
df.fillna(df.isnull().mean(),inplace=True)
```

```python
x=df[['P200SA','P200SL','P200TA','P200TL','P200DA','P200DL','N200TA','N200TL','P3bTA','P3bTL','P3aDA','P3aDL','P50SAvgA','P50TAvgA','P50DAvgA','N100SAvgA','N100TAvgA','N100DAvgA','P200TAvgA','P200DAvgA','N200TAvgA','P3bTAvgA','SlowWaveTAvgA','P3aDAvgA','Missing']]
y=df['Phenotype']
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
le.fit(['HC','AD'])
```

```python
le.transform(['HC','AD'])
```

```python
y=le.transform(y)
```
